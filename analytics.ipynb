```python
# to check missing values in data
sd.isnull().sum()
```

```python
# change missing value with their mean 
sd.groupby('Category')['Review Rating'].mean()
```

```python
sd['Review Rating']=sd.groupby('Category')['Review Rating'].transform(lambda x: x.fillna(x.median()))
```

```python
sd.isnull().sum()
```

```python
# make all columns name in lower case and replace space with underscore
sd.columns=sd.columns.str.lower()
sd.columns=sd.columns.str.replace(" ","_")
sd
```

```python
sd=sd.rename(columns={'purchase_amount_(usd)':'purchase_amount'})
sd
```

```python
# create new column age_group 
label=['Young_Adult','Adult','Middle_aged','Senior']
sd['age_group']=pd.qcut(sd.age,q=4, labels=label)
sd
```

```python
sd[['age','age_group']].head(10)
```

```python
# create column frequency_of_purchases_days (convert words into numbers format).
sd.frequency_of_purchases.unique()
```
